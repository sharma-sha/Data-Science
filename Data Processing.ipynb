#more columns missin value

```python
#fn
def quanQual(dataset):
    qual=[]
    quan=[]
    for column_name in dataset.columns:
        print(column_name)
        if(dataset[column_name].dtype=='O'):
            #print("qual")
            qual.append(column_name)
        else:
            #print("quan")
            quan.append(column_name)
    return quan,qual        
            
```

```python
quan,qaul=quanQual(dataset)
```

```python
dataset[quan]
```

```python
#https://scikit-learn.org/stable/modules/impute.html
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```
