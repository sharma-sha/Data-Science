To calculate the covariance, I'll assume the means of degree_p and etest_p are:

      degree_p: x̄ = 50
      etest_p : ȳ = 150

First, calculate the deviations from the means:

      For X: xi - x̄ = 53.604 - 50 = 3.604
      For Y: yi - ȳ = 176.251 - 150 = 26.251

Then, multiply the deviations:

      (xi - x̄)(yi - ȳ) = 3.604 * 26.251 ≈ 94.63

 the covariance is:
     cov(degree_p, etest_p) ≈ 94.63

Summary:

    - Covariance: 94.634
    - Indicates a positive relationship between degree_p and etest_p
    - As X increases, Y tends to increase
    - Strength of relationship

Here's an example to illustrate the correlation coefficient calculation:

Correlation Coefficient Example

    X = 1.000
    Y = 1.000

Calculate the mean of X (x̄) and Y (ȳ)

    x̄ = 1
    ȳ = 1

Calculate the deviations from the mean

    (xi - x̄) = (1 - 1) = 0
    (yi - ȳ) = (1 - 1) = 0

Calculate the product of the deviations

    (xi - x̄)(yi - ȳ) = 0 * 0 = 0

Calculate the correlation coefficient (corr)

    corr(X, Y) = Σ[(xi - x̄)(yi - ȳ)] / (√Σ(xi - x̄)² * √Σ(yi - ȳ)²)
    = 0 / (√0 * √0)
    = 1 (by definition)

Summary:
        The correlation coefficient is 1.0, indicating a perfect positive linear relationship between X and Y.
        When X increases, Y also increases (perfect positive correlation).
